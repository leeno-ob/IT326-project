#Clustring
In this part, K-means clustering is applied to the preprocessed dataset.  
Different values of K are tested, and the best K is selected using the Silhouette Score and the Elbow Method.

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

url = "https://raw.githubusercontent.com/leeno-ob/IT326-project/refs/heads/main/Dataset/Raw_dataset.csv"
df = pd.read_csv(url)

df.head()
```

```python
# Drop class label for clustering
X = df.drop(columns=['Performance_Level'])

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Shape of X_scaled:", X_scaled.shape)
```

### Choosing the Best K

To determine the optimal number of clusters, we tested different values of K and evaluated them using:
- Silhouette Score
- Total within-cluster sum of squares (WCSS)


```python
k_values = range(2, 6)
silhouette_scores = []
wcss = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)

    silhouette_scores.append(silhouette_score(X_scaled, labels))
    wcss.append(kmeans.inertia_)

    print(f"K = {k}")
    print(f"Silhouette Score = {silhouette_score(X_scaled, labels):.4f}")
    print(f"WCSS = {kmeans.inertia_:.4f}")
    print("-" * 30)
```

```python
plt.figure(figsize=(8,5))
plt.plot(k_values, silhouette_scores, marker='o')
plt.title("Silhouette Score for Different K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.show()
```

### Silhouette Score – Results

The Silhouette Score was calculated for different values of K.  
The best K is the one with the highest Silhouette Score, as it indicates better separation between clusters.


```python
plt.figure(figsize=(8,5))
plt.plot(k_values, wcss, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.show()
```

### Elbow Method – Results

The Elbow Method was used to observe the decrease in WCSS as the number of clusters increases.  
The best K is usually chosen near the point where the decrease becomes less significant.

```python
best_k = k_values[silhouette_scores.index(max(silhouette_scores))]
print("Best K based on Silhouette Score:", best_k)
```

### Best K – Result

The best value of K was selected based on the highest Silhouette Score.  
This value provides the most meaningful and well-separated clusters.

```python
kmeans = KMeans(n_clusters=best_k, random_state=42)
labels = kmeans.fit_predict(X_scaled)

print("First 10 cluster labels:")
print(labels[:10])

print("\nCluster Centers:")
print(kmeans.cluster_centers_)
```

```python
results_df = pd.DataFrame({
    'K': list(k_values),
    'Silhouette Score': silhouette_scores,
    'WCSS': wcss
})

results_df
```

### Clustering Evaluation Table

The following table summarizes the clustering results for each tested value of K, including:
- Silhouette Score
- WCSS

This helps in comparing the clustering quality and selecting the optimal number of clusters.

```python
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-means Clustering Visualization")
plt.show()
```

### Clustering Visualization – Interpretation

The plot shows how the data points are grouped into different clusters based on similarity.  
Each color represents a different cluster identified by the K-means algorithm.  

The clusters appear reasonably separated, which indicates that the model was able to group similar records together.  
Based on the evaluation results, the selected K provides a meaningful clustering result.

### Clustering Findings

K-means clustering was applied using multiple values of K.  
The clustering quality was evaluated using Silhouette Score and WCSS.  

The best K was selected based on the highest Silhouette Score and supported by the Elbow Method.  
The final clustering visualization shows that the dataset can be grouped into meaningful clusters based on similarity between records.
