```python
df.describe()

exam = df['Exam_Score']

min_val = exam.min()
q1 = exam.quantile(0.25)
median = exam.median()
q3 = exam.quantile(0.75)
max_val = exam.max()

print("Minimum:", min_val)
print("Q1:", q1)
print("Median:", median)
print("Q3:", q3)
print("Maximum:", max_val)

iqr = q3 - q1
print("IQR:", iqr)

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = exam[(exam < lower_bound) | (exam > upper_bound)]

print("Number of outliers:", len(outliers))
outliers.head()

import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.boxplot(df['Exam_Score'])
plt.title("Boxplot of Exam Score")
plt.ylabel("Score")
plt.show()

```
